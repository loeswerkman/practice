# Initial analyses
Here the first simple analyses are done to create initial data insights.
##
#### 1. Insights per conversation

```python
import pandas as pd

summaryPerConversation = pd.read_csv('../output/summaryPerConversation.txt', sep="\t", header=None)
summaryPerConversation.columns = ["file", "word_count", "english_word_count", "spanish_word_count", "um_count"]
summaryPerConversation
```

```python
total_word_count = sum(summaryPerConversation.word_count)
total_um_count = sum(summaryPerConversation.um_count)

print(
    f'Number of recorded conversations: { len(summaryPerConversation.index) }'
    f'\nTotal word count: { total_word_count }'
    f'\nTotal um count: { total_um_count }'
    f'\nPercentage of um: { ( total_um_count / total_word_count ) * 100 }'
)
```

##
#### 2. Insights per person

```python
import pandas as pd

summaryPerPerson = pd.read_csv('../output/summaryPerPerson.txt', sep="\t", header=None)
summaryPerPerson.columns = ["speaker", "word_count", "english_word_count", "spanish_word_count", "um_count"]
summaryPerPerson
```

```python
print(
    f'Number of recorded speakers: { len(summaryPerPerson.index) }'
)
```

```python
import matplotlib.pyplot as plt

plt.boxplot(summaryPerPerson.word_count)
plt.title('Distribution of total word count per person')
plt.ylabel('number of spoken words')
plt.show()
```

```python
plt.boxplot(summaryPerPerson.um_count)
plt.title('Distribution of total um count per person')
plt.ylabel('total utterances of "um"')
plt.show()
```

```python
fig = plt.figure(figsize=(15,4))
ax = fig.add_axes([0,0,1,1])
summaryPerPerson.sort_values("word_count")
ten_most_frequent_speakers = summaryPerPerson.head(10)
ax.bar(ten_most_frequent_speakers.speaker, ten_most_frequent_speakers.um_count/ten_most_frequent_speakers.word_count)
plt.xticks(fontsize=12)
plt.show()
```
