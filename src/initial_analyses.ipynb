# Initial analyses
Here the first simple analyses are done to create initial data insights.
##
#### 1. Insights per conversation

```python
import pandas as pd

summary = pd.read_csv('../output/summaryPerPersonPerConversation.txt', sep="\t")
summary
```

```python
summaryPerConversation = summary.groupby('conversation').sum()
summaryPerConversation
```

```python
print(
    f'Number of recorded conversations: {len(summaryPerConversation.index)}'
    f'\nTotal word count: {sum(summaryPerConversation.total)}'
    f'\nTotal um count: {sum(summaryPerConversation.um)}'
    f'\nTotal yeah count: {sum(summaryPerConversation.yeah)}'
    f'\nTotal like count: {sum(summaryPerConversation.like)}'
    f'\nTotal pues count: {sum(summaryPerConversation.pues)}'
    f'\nTotal eso count: {sum(summaryPerConversation.eso)}'
)
```

##
#### 2. Insights per person

```python
summaryPerPerson = summary.groupby('speaker').sum()
summaryPerPerson
```

```python
print(
    f'Number of recorded speakers: { len(summaryPerPerson.index) }'
)
```

```python
import matplotlib.pyplot as plt

plt.boxplot(summaryPerPerson.total)
plt.title('Distribution of total word count per person')
plt.ylabel('number of spoken words')
plt.show()
```

```python
plt.boxplot(summaryPerPerson.um)
plt.title('Distribution of total um count per person')
plt.ylabel('total utterances of "um"')
plt.show()
```
