# Initial analyses
Here the first simple analyses are done to create initial data insights.

##
#### 2. Insights per person

```python
import pandas as pd

summaryPerPerson = pd.read_csv('../output/summaryPerPerson.txt', sep="\t")
summaryPerPerson
```

```python
print(
    f'Number of recorded speakers: { len(summaryPerPerson.index) }'
)
```

```python
import matplotlib.pyplot as plt

plt.boxplot(summaryPerPerson.total)
plt.title('Distribution of total word count per person')
plt.ylabel('number of spoken words')
plt.show()
```

```python
plt.boxplot(summaryPerPerson.um)
plt.title('Distribution of total um count per person')
plt.ylabel('total utterances of "um"')
plt.show()
```

```python
fig = plt.figure(figsize=(15,4))
ax = fig.add_axes([0,0,1,1])
summaryPerPerson.sort_values("total")
ten_most_frequent_speakers = summaryPerPerson.head(10)
ax.bar(ten_most_frequent_speakers.name, ten_most_frequent_speakers.um/ten_most_frequent_speakers.total)
plt.xticks(fontsize=12)
plt.show()
```
